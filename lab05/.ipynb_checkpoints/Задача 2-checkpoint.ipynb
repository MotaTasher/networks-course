```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
u_1 = 300
u_2 = 700
u_3 = 2 * 10 ** 3
F = 15 * 10 ** 6
u_s = 30 * 10 ** 3
d = 2 * 10 ** 3
```

```python
def one_range_architecture(F, d, u_s, N, u):
    return np.maximum(max(F / u_s, F / d), N * F / (u_s + N * u))

def server_client_architecture(F, d, u_s, N, u):
    return np.maximum(N * F/ u_s, F / d)
```

```python
bins = np.arange(1, 1000)
plt.xlabel('N (Кб)')
plt.ylabel('D', rotation=0)
plt.plot(bins, one_range_architecture(F, d, u_s, bins, u_1), label='u = 300 Кбит/с')
plt.plot(bins, one_range_architecture(F, d, u_s, bins, u_2), label='u = 700 Кбит/с')
plt.plot(bins, one_range_architecture(F, d, u_s, bins, u_3), label='u = 2 Мбит/с')
plt.plot(bins, server_client_architecture(F, d, u_s, bins, u_1), label='клиент-сервер')
plt.legend()
plt.show()
```

```python
bins = np.arange(1, 100)
plt.xlabel('N')
plt.ylabel('D', rotation=0)
plt.plot(bins, one_range_architecture(F, d, u_s, bins, u_1), label='u = 300 Кбит/с')
plt.plot(bins, one_range_architecture(F, d, u_s, bins, u_2), label='u = 700 Кбит/с')
plt.plot(bins, one_range_architecture(F, d, u_s, bins, u_3), label='u = 2 Мбит/с')
plt.plot(bins, server_client_architecture(F, d, u_s, bins, u_1), label='клиент-сервер')
plt.legend()
plt.show()
```
